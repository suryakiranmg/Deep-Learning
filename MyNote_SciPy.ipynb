### <font color=green>Operates on numpy arrays</font>

```python
# ----- Generating random Vars in Python ----- #
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0) # For reproducibility

# Using NumPy

samples = np.random.normal(size=1000)
num_bins = 50
plt.hist(samples,50,color='blue')
plt.title('Univariate Normal Dist using NumPy')
plt.show()

# Using SciPy

samples = scipy.stats.norm.rvs(size=1000)
num_bins = 50
plt.hist(samples,50,color='blue')
plt.title('Univariate Normal Dist using SciPy')
plt.show()
```
