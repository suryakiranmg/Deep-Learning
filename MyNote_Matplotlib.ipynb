## <font color=red>Matplotlib is a python package to do 2-D Graphics </font>
### **<font color=blue>Pyplot provides a convenient interface to matplotlib</font>**
### **<font color=blue>Modeled after MATLAB, works with NumPy objects</font>**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi,2*np.pi,256,endpoint=True)
C, S = np.cos(x), np.sin(x)
plt.figure(figsize=(8,5))
plt.plot(x, C, 'b--', linewidth=2.0, label='Cosine')
plt.plot(x, S, 'r--', linewidth=2.0, label='Sine')
plt.legend(loc='lower left')
plt.title('Sine & Cosine')
plt.show()
```
